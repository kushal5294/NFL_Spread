```python
import pandas as pd

# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Prompt the user to enter the names of CSV files
#file_names = ['2016_all_stats.csv', '2017_all_stats.csv', '2018_all_stats.csv', '2019_all_stats.csv',
            #  '2020_all_stats.csv', '2021_all_stats.csv', '2022_all_stats.csv'] 

file_names = ['2016_all_stats.csv', '2017_all_stats.csv'] 
# Loop through the entered CSV file names, read each file into a DataFrame, and append it to the combined_df
for file_name in file_names:
    try:
        data = pd.read_csv(file_name)
        combined_df = pd.concat([data, combined_df], axis=0)
    except FileNotFoundError:
        print(f"File '{file_name}' not found. Skipping...")

# Reorder the columns so that "Tm" is the first column
if 'Tm' in combined_df.columns:
    columns = ['Tm'] + [col for col in combined_df if col != 'Tm']
    combined_df = combined_df[columns]

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combined_data.csv', index=False)

print("Combined data saved to 'combined_data.csv'")
```
